```python
import numpy as np
import pandas as pd

import plotly.express as px
import dash
import dash_html_components as html
import dash_core_components as dcc
import dash_table
from dash.dependencies import Input, Output, State
```

```python
df = px.data.gapminder()
radio_item = dcc.RadioItems(
            id = 'radio-item',
            options = [{'label':i,'value':i} for i in  df.continent.unique()]
        )

drop_down = dcc.Dropdown(
            id = 'drop-down'
        )
slider =  dcc.RangeSlider(
            id = 'slider',
            min = df.year.min(),
            max = df.year.max(),
            marks = {int(i):str(i)+'년도'  for i in df.year.unique()}          
        )

bar_chart = dcc.Graph(
            id = 'bar-chart'
        )

data_table = dash_table.DataTable(
            id = 'table'
        )

input_text = dcc.Input(
            id = 'input-text',
            type = 'text'
        )

drop_down_table  = dcc.Dropdown(
                      id = 'drop-down-table',
                    options = [{'label':i,'value':i} for i in df.columns],
                    multi = True
                )
```

```python
app = dash.Dash()
app.layout = html.Div(
    [
        html.Label('대륙'),
        radio_item,
        html.Label('국가'),
        drop_down,
        html.Label('년도'),
        slider,
        bar_chart,
        input_text,
        drop_down_table,
        html.Button('클릭',
                    id = 'submit-val',
                    n_clicks = 0
                   ),
        data_table
    ]
)
```

```python
@app.callback(
    Output('drop-down','options'),
    Input('radio-item','value')
)

def update_country_dropdown(continent):
    option = [{'label':i,'value':i} for i in df.loc[df.continent==continent,'country'].unique()]
    return option


@app.callback(
    Output('bar-chart','figure'),
    Input('drop-down','value'),
    Input('slider','value')
)

def update_bar_chart(country,year):
    data = df.loc[(df.country==country) & ((df.year >= int(year[0])) & (df.year <= int(year[1]))),['year','pop']]
    fig = px.bar(data,x='year',y='pop',color='pop')
    return fig

@app.callback(
    Output('table','columns'),
    Output('table','data'),
    Input('submit-val','n_clicks'),
    State('input-text','value'),
    State('drop-down-table','value')
)
def update_table(n,country,cols):
    columns = [{'name':i,'id':i} for i in cols]
    data = df.loc[df.country.str.contains(country),cols].to_dict('records')
    return columns, data



```

```python
app.run_server(port='7777')
```
